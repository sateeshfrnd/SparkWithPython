## Compare DataFrame and get all the rows of DataFrame that do not exist in another DataFrame

```
df1 = spark.createDataFrame(
  [('Shiva', 'Appu'), ('Teja', 'Sweety'), ('Bhavishya', 'Bhavi')],
  ['name','nick_name']
)
df1.show(truncate=False)
```

```
df2 = spark.createDataFrame(
  [('Shiva', 'Chinu'), ('Teja', 'Sweety'), ('Bhavishya', 'Bhavi')],
  ['name','nick_name']
)
df2.show(truncate=False)
```

### Get all rows in df1 that does not exist in df2

```
df1.exceptAll(df2).show()
```
