## Partitioning a DataFrame
describe(~) method returns a new PySpark DataFrame holding summary statistics of the specified columns..

### Parameters
*cols | string | optional
- By default, all numeric and string columns will be described.

### Return Value
  - A new PySpark DataFrame.

[API Reference](https://spark.apache.org/docs/3.1.1/api/python/reference/api/pyspark.sql.DataFrame.describe.html)

```
df = spark.createDataFrame(
  [('Shiva', 1), ('Teja', 7), ('Bhavishya', 5)],
  ['name','age']
)
df.show(truncate=False)
```

### Getting summary statistics of certain columns in the DataFrame

```
df.describe("name","age").show()
```

### Getting summary statistics of all numeric and string columns in the DataFrame

```
df.describe().show()
```
